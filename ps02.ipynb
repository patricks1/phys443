```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import rcParams
```

```python
rcParams['mathtext.fontset'] = 'dejavuserif'
rcParams['font.family'] = 'serif' 
rcParams['xtick.labelsize'] = 16
rcParams['ytick.labelsize'] = 16
rcParams['axes.grid']=True
rcParams['axes.titlesize']=24
rcParams['axes.labelsize']=20
rcParams['axes.titlepad']=15
rcParams['legend.frameon'] = True
rcParams['legend.facecolor']='white'
rcParams['legend.fontsize']=18
```

```python
dtype=np.dtype({'names':['ra','dec','u','g','r','z','typ'],
                'formats':[np.float,np.float,np.float,
                           np.float,np.float,np.float,
                           'S2']})
ra,dec,u,g,r,z,typ=np.loadtxt('dat.csv',skiprows=1,usecols=[2,3,4,5,6,8,10],
                          unpack=True,delimiter=',',dtype=dtype)
```

```python
print min(z)
print max(z)
```

```python
ug=u-g
gr=g-r
isel=typ=='el'
issp=typ=='sp'

fig=plt.figure(figsize=(8,6))
ax=fig.add_subplot(111)
ax.plot(ug[isel],gr[isel],'rd',label='elliptical')
ax.plot(ug[issp],gr[issp],'bd',label='spirals')
ax.legend()
ax.set_xlabel('u-g')
ax.set_ylabel('g-r')
ax.set_title('Color-color diagram, Abell 2255')
plt.show()
```

### 12" radius

```python
typ_b,u_b,g_b,r_b,z_b=np.loadtxt('dat_big.csv',delimiter=',',skiprows=2,
                             usecols=[6,9,10,11,13],unpack=True)

#z filtered
isgal=typ_b==3
inz=(z_b<19) & (z_b>14)
inc=isgal&inz
ug_b=u_b-g_b
gr_b=g_b-r_b

fig=plt.figure(figsize=(8,6))
ax=fig.add_subplot(111)
ax.plot(ug_b[inc],gr_b[inc],'gd',alpha=0.6,ms=4)
ax.legend()
ax.set_xlabel('u-g')
ax.set_ylabel('g-r')
ax.set_title('Color-color diagram, Abell 2255',pad=40)
ax.text(0.4,1.05,'$z\in(14,19)$',fontsize=20,transform=ax.transAxes)
ax.set_xlim(-0.5,3.)
ax.set_ylim(-0.5,2.)
plt.show()

#not z filtered
inc=isgal
fig=plt.figure(figsize=(8,6))
ax=fig.add_subplot(111)
ax.plot(ug_b[inc],gr_b[inc],'gd',alpha=0.6,ms=4)
ax.legend()
ax.set_xlabel('u-g')
ax.set_ylabel('g-r')
ax.set_title('Color-color diagram, Abell 2255',pad=40)
ax.text(0.4,1.05,'all $z$',fontsize=20,transform=ax.transAxes)
ax.set_xlim(-0.5,3.)
ax.set_ylim(-0.5,2.)
plt.show()
```
